# 03 — Model conversion and LoRA merge
Use a versioned `ModelTemplate`; never guess a foreign checkpoint layout.

```python
import numpy as np
from tunix_craftext.interop import LoraAdapter, ModelTemplate, TensorRule, convert_state_dict, merge_lora_adapters
```

```python
external = {'linear.weight': np.arange(6, dtype=np.float32).reshape(3, 2), 'linear.bias': np.zeros(3, dtype=np.float32)}
template = ModelTemplate('tiny-linear', (TensorRule('linear.weight', ('Dense_0', 'kernel'), 'transpose_2d', (2, 3)), TensorRule('linear.bias', ('Dense_0', 'bias'), expected_shape=(3,))))
params = convert_state_dict(external, template)
params
```

```python
adapter = LoraAdapter(target=('Dense_0', 'kernel'), down=np.array([[1.0, 0.0]]), up=np.array([[1.0], [2.0], [3.0]]), alpha=1.0)
merged = merge_lora_adapters(params, [adapter])
print('base kernel:\n', params['Dense_0']['kernel'])
print('merged kernel:\n', merged['Dense_0']['kernel'])
```

For a new architecture, commit the template, golden source tensors and an output-parity test together.
