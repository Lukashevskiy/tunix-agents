# 04 — MegaPrompts: environment state → model prompt
This notebook uses the vendored MegaPrompts `base` template. Install the local optional dependency first: `pyenv exec python -m uv sync --extra examples --extra prompts`.


```python
from types import SimpleNamespace
import numpy as np
from tunix_craftext import ActionCatalog, MegaPromptRenderer, PromptContext
```

The real Craftax state has these fields. A small state-shaped object lets us inspect prompt construction without starting a full environment.


```python
observation = SimpleNamespace(
    map=np.zeros((3, 3), dtype=np.int32),
    player_position=np.array([1, 1]),
    player_direction=np.array(1),
    inventory={'wood': 2},
)
actions = ActionCatalog(('LEFT', 'RIGHT', 'DO'))
context = PromptContext(goal='collect wood', observation=observation, actions=actions, safety='Do not enter lava.')
```

```python
rendered = MegaPromptRenderer('base').render(context)
print(rendered.text)
print('Action id for DO:', rendered.actions.index_of('DO'))
```

```python
# A model decoder must use the catalogue from *this exact rendered prompt*.
model_label = 'DO'
environment_action_id = rendered.actions.index_of(model_label)
environment_action_id
```

Unknown labels raise `PromptContractError`; they are never silently coerced. The next notebook/module will replace `model_label` with a typed `TextPolicy` output and record invalid-action metrics.
