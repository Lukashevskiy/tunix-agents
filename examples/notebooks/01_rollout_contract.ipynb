# 01 — Trajectory contract and rollout
Build a tiny policy/environment and inspect a time-major `[T, B, ...]` rollout.

```python
import numpy as np
from tunix_craftext import collect_rollout
```

```python
def policy(observation):
    return observation + 1, np.full(observation.shape, -0.5), np.zeros(observation.shape)

def step(state, action):
    next_state = state + action
    return next_state, next_state, action.astype(float), next_state >= 6, np.zeros_like(next_state, dtype=bool)
```

```python
state, observation, rollout = collect_rollout(np.array([0, 0]), np.array([0, 0]), 3, policy, step)
print('reward [T, B]:', rollout.transitions.reward.shape)
print('done [T, B]:', rollout.transitions.done)
print('bootstrap [B]:', rollout.bootstrap_value)
```
